```python
# prompt: create deployment dynamic colorful graph  for displaying each in graph 'AgeGroup', 'Attrition', 'BusinessTravel', 'DailyRate','Department', 'DistanceFromHome', 'Education', 'EducationField', 'EnvironmentSatisfaction', 'Gender','HourlyRate', 'JobInvolvement', 'JobLevel', 'JobRole','JobSatisfaction', 'MaritalStatus', 'MonthlyIncome', 'SalarySlab','MonthlyRate',

import plotly.express as px

def plot_distribution(df, column):
  """
  Plots a dynamic and colorful distribution graph for a given column.

  Args:
    df: The Pandas DataFrame containing the data.
    column: The name of the column to plot.
  """
  fig = px.histogram(df, x=column, color=column,
                     title=f"Distribution of {column}",
                     labels={column: column},
                     marginal="box",  # Add a box plot on the right
                     opacity=0.7,
                     template="plotly_white")
  fig.update_layout(bargap=0.1)
  fig.show()

# List of columns to plot
columns_to_plot = ['AgeGroup', 'Attrition', 'BusinessTravel', 'DailyRate',
                   'Department', 'DistanceFromHome', 'Education', 'EducationField',
                   'EnvironmentSatisfaction', 'Gender', 'HourlyRate', 'JobInvolvement',
                   'JobLevel', 'JobRole', 'JobSatisfaction', 'MaritalStatus',
                   'MonthlyIncome', 'SalarySlab', 'MonthlyRate']

# Plot the distributions
for column in columns_to_plot:
  plot_distribution(df, column)
```

```python
# prompt: create deployment dynamic graph which have selected by attration for displaying each in graph 'NumCompaniesWorked', 'OverTime','PercentSalaryHike', 'PerformanceRating', 'RelationshipSatisfaction','StandardHours', 'StockOptionLevel', 'TotalWorkingYears','TrainingTimesLastYear', 'WorkLifeBalance', 'YearsAtCompany','YearsInCurrentRole', 'YearsSinceLastPromotion','YearsWithCurrManager'

import plotly.express as px
#Import the ipywidgets module
import ipywidgets as widgets

def create_deployment_graph(selected_attribute="NumCompaniesWorked"):
  """
  Creates a deployment graph based on the selected attribute,
  specifically for attrition analysis.

  Args:
      selected_attribute (str): The attribute to display in the graph.
                                Defaults to "NumCompaniesWorked".
  """
  fig = px.histogram(df, x=selected_attribute, color="Attrition",
                     barmode='group', title=f"Employee Deployment by Attrition - {selected_attribute}")
  fig.update_layout(
      xaxis_title=selected_attribute,
      yaxis_title="Employee Count",
      legend_title="Attrition",
      height=500,
      width=800
  )
  fig.show()

# List of attributes relevant to attrition analysis
attributes = ['NumCompaniesWorked', 'OverTime','PercentSalaryHike', 'PerformanceRating',
              'RelationshipSatisfaction','StandardHours', 'StockOptionLevel',
              'TotalWorkingYears','TrainingTimesLastYear', 'WorkLifeBalance',
              'YearsAtCompany','YearsInCurrentRole', 'YearsSinceLastPromotion',
              'YearsWithCurrManager']

# Create interactive dropdown for attribute selection (not supported in Colab)
selected_attribute = "NumCompaniesWorked"  # Default selection

# Create the initial graph
create_deployment_graph(selected_attribute)

# Add dropdown widget (not supported in Colab, but you can adapt this for other environments)
widgets.interact(create_deployment_graph, selected_attribute=attributes)
```

## Attrition

```python
# prompt: buatkan prediksi penurunan jumlah pegawai pertahun dan sajikan sisa jumlah pegawai per tahun  berdasarkan average attrition rate yang ada dan buatkan grafiknya tampilkan jumlahnya

# Calculate average attrition rate
average_attrition_rate = df['Attrition'].value_counts(normalize=True)['Yes']

# Get initial employee count
initial_employees = len(df)

# Project employee count for the next 10 years
years = list(range(1, 11))
projected_employees = []
remaining_employees = []

for year in years:
  attrition_count = round(initial_employees * average_attrition_rate)
  remaining_employees.append(initial_employees - attrition_count)
  projected_employees.append(attrition_count)
  initial_employees -= attrition_count

# Create a DataFrame for plotting
projection_df = pd.DataFrame({
    'Year': years,
    'Attrition Count': projected_employees,
    'Remaining Employees': remaining_employees
})

# Plot the projection
fig = px.line(projection_df, x='Year', y=['Attrition Count', 'Remaining Employees'],
              title='Employee Attrition Projection')
fig.update_layout(
    xaxis_title='Year',
    yaxis_title='Employee Count',
    legend_title='Legend'
)
fig.show()

# Display the projection DataFrame
print(projection_df)
```

## Attrition Key Observations:


*   Year-on-Year Decrease:  Each year, the number of employees leaving the company is projected to remain relatively consistent,leading to a steady decline in the total workforce.List item
*   Cumulative Impact: Over the 10-year period, the cumulative effect of attrition could result in a substantial reduction in the company's employee base. This could have significant

implications for productivity, knowledge retention, and overall business continuity.
